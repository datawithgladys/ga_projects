To gain a fuller understanding on the trends and factors influencing participation rates of the SAT in the US, an analysis of various states in the US regarding the following has been made:

- the participation rates of each state

- any other policies (e.g. subsidies) implemented with regards to college admission exams in each state

With insights made from these analyses, this project aims to suggest a suitable state and recommend actions to be taken to improve the state's SAT participation rate moving forward.

#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#function to calculate standard deviation for a series or column of dataframe

def standard_deviation(column):
    std = 0
    mean = column.mean() #work out the mean
    for value in range(len(column)):
        std += (column[value] - mean) ** 2 #subtract the mean from each value, and square it
    std = math.sqrt(std/len(column)) #divide above value with the total number of values in the column, and then square root it
    return std
````

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels):
    nrows = int(np.ceil(len(list_of_columns)/2)) # Ensures enough rows are given for histogram
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize = (10,10), sharey=True) #specify figsize
    ax = ax.ravel()
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column])
        # Set titles, labels, etc here for each subplot
        ax[i].set_title(list_of_titles[i], fontsize= 12)
        ax[i].set_xlabel(list_of_xlabels[i],fontsize= 8)
        ax[i].set_ylabel('Total number of states',fontsize= 8)
````

#### Lower-income students in West Virginia

States such as West Virginia has been struggling to overcome devastating cycles of poverty and “brain drain" in lower-income students located in their respective states, which stems from low educational attainment rates and the absence of a college-going culture <sup>[[14]](http://www.wvhepc.edu/wp-content/uploads/2014/09/RFP-15045-Appendix-1-Master-grant-document-Abridged-version.pdf)</sup>. This is clearly evident by the lower exam participation rates for both tests in West Virginia during 2017 and 2018. With an absence of a college-going culture for the lower-income families, this negatively affects the participation rates for any college admission exams.

In contrast, the ACT test costs around the same at 55 US dollars <sup>[[16]](http://www.act.org/content/act/en/products-and-services/the-act/registration/fees.html)</sup>. Even though some form of testing-fee waivers are available for low-income students, they typically are underused. Nationwide, over half of students nationwide are eligible for an SAT fee waiver. Yet just 25% and 21% of students who took the SAT in 2017 used a waiver <sup>[[17]](https://www.nytimes.com/2017/07/14/upshot/how-universal-college-admission-tests-help-low-income-students.html)</sup>. In fact, the College Board estimates about one-fourth of eligible students don’t take advantage of the fee waivers they're entitled to <sup>[[18]](https://www.collegeboard.org/releases/2018/more-than-2-million-students-in-class-of-2018-took-sat-highest-ever)</sup>.

However, the College Board has been trying to increase SAT participation through fee waivers for low-income students, and this clearly is working to a certain extent for states like West Virginia, contributing to an increase in 14% participation rate for the SAT in 2018 without any mandatory SAT testing required in 2018.
